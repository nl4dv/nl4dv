# Imports

```python
import os
from nl4dv import NL4DV
```

# Initializing NL4DV with a Housing Dataset

```python
nl4dv_instance = NL4DV(data_url=os.path.join("..","..","assets","data","housing.csv"))
dependency_parser_config = {'name': 'corenlp','model': os.path.join("..","..","assets","jars","stanford-english-corenlp-2018-10-05-models.jar"),'parser': os.path.join("..","..","assets","jars","stanford-parser.jar")}
nl4dv_instance.set_dependency_parser(config=dependency_parser_config)
```

# Rendering the Most Relevant Visualization


```python
nl4dv_instance.render_vis("Show average prices for different home types over the years")
```

# Auto-detecting Follow Up Queries

```python
nl4dv_instance.render_vis("As a bar chart", dialog='auto')
```

# Generating a New Standalone Visualization

```python
nl4dv_instance.render_vis("Correlate Price and Lot Area", dialog=False)
```

#  Following Up on a Previous Conversation

```python
nl4dv_instance.render_vis("Just show condos and duplexes", dialog=True, dialog_id = "0", query_id = "1")
```
